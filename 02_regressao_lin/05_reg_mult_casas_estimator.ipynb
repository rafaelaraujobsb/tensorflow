```python
import pandas as pd
import tensorflow as tf
```

```python
df = pd.read_csv('../data/house-prices.csv')
df.head()
```

### Pré-processamento

```python
colunas_usadas = df.columns[2:-2]
colunas_usadas
```

```python
df = df.loc[:, df.columns.isin(colunas_usadas)]
df.head()
```

#### Escalonamento

```python
from sklearn.preprocessing import MinMaxScaler # normalização, valores entre 0 e 1
```

```python
scaler_x = MinMaxScaler()
df[df.columns[1:]] = scaler_x.fit_transform(df[df.columns[1:]])
df.head()
```

```python
scaler_y = MinMaxScaler()
df[['price']] = scaler_y.fit_transform(df[['price']])
df.head()
```

#### Classe / Previsores

```python
X = df.drop('price', axis=1)
y = df.price
```

```python
X.head()
```

```python
y.head()
```

#### feature_column

```python
colunas = [ tf.feature_column.numeric_column(key=c) for c in colunas_usadas[1:]]
colunas
```

#### Split base

```python
from sklearn.model_selection import train_test_split
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(X, y, test_size=.3)
X_treinamento.shape, X_teste.shape, y_treinamento.shape, y_teste.shape
```

### Treinamento

```python
funcao_treinamento = tf.estimator.inputs.pandas_input_fn(
    x=X_treinamento,
    y=y_treinamento,
    batch_size=32,
    num_epochs=None,
    shuffle=True
)
```

```python
funcao_teste = tf.estimator.inputs.pandas_input_fn(
    x=X_treinamento,
    y=y_treinamento,
    batch_size=32,
    num_epochs=10000,
    shuffle=False
)
```

```python
regressor = tf.estimator.LinearRegressor(feature_columns=colunas)
```

```python
regressor.train(input_fn=funcao_treinamento, steps=10000)
```

### Métricas

```python
metricas_treinamento = regressor.evaluate(input_fn=funcao_treinamento, steps=10000)
```

```python
metricas_teste = regressor.evaluate(input_fn=funcao_teste, steps=10000)
```

```python
metricas_treinamento
```

```python
metricas_teste
```

### Predict

```python
funcao_previsao = tf.estimator.inputs.pandas_input_fn(
    x=X_teste, 
    shuffle=False
)
```

```python
previsoes = [ p['predictions'] for p in regressor.predict(input_fn=funcao_previsao)]
```

```python
previsoes[:5]
```

### Mean Abs Error

```python
import numpy as np
from sklearn.metrics import mean_absolute_error
```

```python
valores_previsoes = np.asarray(previsoes).reshape(-1,1)
```

```python
valores_previsoes = scaler_y.inverse_transform(valores_previsoes)
valores_previsoes[:5]
```

```python
y_teste2 = scaler_y.inverse_transform(y_teste.values.reshape(-1,1))
y_teste2[:5]
```

```python
mean_absolute_error(y_teste2, valores_previsoes)
```
